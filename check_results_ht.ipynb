```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, auc
from IPython.display import clear_output
import os
import shutil
import re
from datetime import datetime, timedelta
import seaborn as sns
from IPython.display import Image, display, IFrame
from jDataResultsAnalysis import explore_data_results, compare_data_results, explore_optuna_results
%matplotlib inline
```

```python
clinical_data = pd.read_csv("ctg_dataset/Old Dataset/archive/clinical_data.csv")
X = np.load("ctg_dataset/Old Dataset/archive/X.npy")
y = np.load("ctg_dataset/Old Dataset/archive/y.npy")

print(clinical_data.shape, X.shape, y.shape)

# Split data into FHR (X[:,:,0]) and TOCO (X[:,:,1]) components
X_fhr = X[:, :, 0]  # FHR signal
X_toco = X[:, :, 1]  # TOCO signal

# Split the dataset into training and validation sets (80% train, 20% val)
X_fhr_train, X_fhr_val, X_toco_train, X_toco_val, y_train, y_val, clinical_train, clinical_val = train_test_split(
    X_fhr, X_toco, y, clinical_data, test_size=0.2, random_state=42, stratify=y)

print(X_fhr_train.shape, X_toco_train.shape, y_train.shape, clinical_train.shape)
print(X_fhr_val.shape, X_toco_val.shape, y_val.shape, clinical_val.shape)

# Save the split datasets into the desired files
np.save("ctg_dataset/Old Dataset/X_train_fhr.npy", X_fhr_train)
np.save("ctg_dataset/Old Dataset/X_val_fhr.npy", X_fhr_val)
np.save("ctg_dataset/Old Dataset/X_train_toco.npy", X_toco_train)
np.save("ctg_dataset/Old Dataset/X_val_toco.npy", X_toco_val)
np.save("ctg_dataset/Old Dataset/y_train.npy", y_train)
np.save("ctg_dataset/Old Dataset/y_val.npy", y_val)

# Save clinical data to CSV files
clinical_train.to_csv("ctg_dataset/Old Dataset/clinical_train.csv", index=False)
clinical_val.to_csv("ctg_dataset/Old Dataset/clinical_val.csv", index=False)

print("Data reorganization and split completed successfully.")
```
